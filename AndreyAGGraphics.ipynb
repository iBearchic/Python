```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import numpy as np
from numpy import sin, cos, pi
```

```python
r = 1
zetta = np.linspace(0, 2 * pi, 100)
phi= sin(zetta/2)

x = r * (sin(phi)) * (sin(zetta))
y = r * (sin(phi)) * (sin(zetta))
z = r * (cos(phi))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='parametric curve')

plt.title("Сферическая С.К.", color = 'red')
ax.set_xlabel('ОСЬ Х', fontsize = 10, color = 'red')
ax.set_ylabel('ОСЬ У', fontsize = 10, color = 'red')

plt.show()
```

```python
zetta = np.linspace(0, 2 * pi, 100)

rx = 1 + sin (3*zetta)

x = rx *cos(zetta)
y = rx * sin(zetta)
z = cos(zetta)




fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='parametric curve')

plt.title("Цилиндрическая С.К.", color = 'red')
ax.set_xlabel('ОСЬ Х', fontsize = 10, color = 'red')
ax.set_ylabel('ОСЬ У', fontsize = 10, color = 'red')

plt.show()
```

```python
rx1=1
rx2=2

zetta1 = np.linspace(0, 6*pi, 100)
zetta2 = pi + 1/3 * zetta1

x = rx1 * cos(zetta1)
y = rx1 * sin(zetta1)

fig,ax = plt.subplots()

ax.plot(x, y)

x = rx2 * cos(zetta2)
y = rx2 * sin(zetta2)

ax.plot(x, y)

plt.title("Солнечная система", color = 'red')
ax.set_xlabel('ОСЬ Х', fontsize = 10, color = 'red')
ax.set_ylabel('ОСЬ У', fontsize = 10, color = 'red')

ax.grid()

plt.show()
```
